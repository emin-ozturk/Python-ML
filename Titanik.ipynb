```python
import seaborn as sns
import pandas as pd
```

# Veri seti yüklenmesi

```python
titanic = sns.load_dataset("titanic")
titanic.head(5)
```

```python
titanic.shape
```

# Boş değerlerin belirlenmesi ve silinmesi

```python
titanic.isna().sum() / len(titanic)
```

```python
columns_names = [col for col in titanic.columns if titanic[col].isna().sum() / len(titanic) < 0.7]
columns_names
```

```python
df = titanic[columns_names]
```

# Benzer olan sütunların silinmesi

```python
df.drop(["alive", "embarked", "adult_male", "sibsp", "who"], axis=1, inplace=True)
df.head(5)
```

# Guruplandırma ve kırılım inceleme

```python
df.groupby("pclass").agg({"fare": ["max", "min", "mean", "median"]})
```

```python
df.groupby(["embark_town", "sex"]).agg({"survived": "mean"})
```

```python
df.groupby(["pclass", "alone"]).agg({"survived": "mean"})
```

```python
agg_df = df.groupby(["class", "sex", "embark_town"]).agg({"survived": "mean", "age": "mean"})
agg_df
```

```python
agg_df.reset_index(inplace = True)
```

```python
agg_df
```

```python
agg_df.shape
```

```python
agg_df["person_info"] = agg_df[["class", "sex", "embark_town"]].agg(lambda x: "_".join(x).upper(), axis=1)
```

```python
agg_df.head()
```
