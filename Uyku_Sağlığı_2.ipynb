```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score

pd.set_option('display.max_column', None)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/emin-ozturk/Python-ML/main/Uyku%20Sa%C4%9Fl%C4%B1%C4%9F%C4%B1/Sleep_health_and_lifestyle_dataset.csv")
df.head(7)
```

```python
blood_pressure = df["Blood Pressure"]
df[["High Blood Pressure", "Low Blood Pressure"]] = blood_pressure.str.split("/", expand=True)
df["High Blood Pressure"] = df["High Blood Pressure"].astype(int)
df["Low Blood Pressure"] = df["Low Blood Pressure"].astype(int)
```

```python
df.drop(["Person ID", "Blood Pressure"], axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.head()
```

```python
cat_col = [col for col in df.columns if str(df[col].dtype) in ["object"]]
cat_col
```

```python
num_col = [col for col in df.columns if df[col].dtype in ["int", "float"]]
num_col
```

```python
cat_col_encoding = pd.DataFrame()

for col in cat_col:
  if col != "Sleep Disorder":
    cat_col_dummies = pd.get_dummies(df[col])
    cat_col_encoding = pd.concat([cat_col_encoding, cat_col_dummies], axis = 1)

cat_col_encoding.head(10)
```

```python
encoding_df = pd.concat([df[num_col], cat_col_encoding, df["Sleep Disorder"]], axis = 1)
encoding_df.head(10)
```

```python
encoding_df.shape
```

```python
x = encoding_df.iloc[:, 0:-1]
y = encoding_df.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
lg = DecisionTreeClassifier()
lg.fit(x_train, y_train)
y_pred = lg.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc}")
print(f"Confusion_Matrix:\n{cm}")

gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc}")
print(f"Confusion_Matrix:\n{cm}")
```
