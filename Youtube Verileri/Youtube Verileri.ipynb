# **Boş değer doldurma**

```python
print(dataset.isna().sum())
```

```python
print(dataset[["category"]].isna().sum())
dataset['category'] = dataset['category'].fillna(dataset['category'].mode()[0])
print(dataset[["category"]].isna().sum())
```

```python
print(dataset[["Abbreviation"]].isna().sum())
dataset['Abbreviation'] = dataset['Abbreviation'].fillna(dataset['Abbreviation'].mode()[0])
print(dataset[["Abbreviation"]].isna().sum())
```

```python
print(dataset[["channel_type"]].isna().sum())
dataset['channel_type'] = dataset['channel_type'].fillna(dataset['channel_type'].mode()[0])
print(dataset[["channel_type"]].isna().sum())
```

```python
print(dataset[["video_views_rank"]].isna().sum())
dataset['video_views_rank'] = dataset['video_views_rank'].fillna(dataset['video_views_rank'].mean())
print(dataset[["video_views_rank"]].isna().sum())
```

```python
for i in dataset:
  if dataset[i].isna().sum() != 0:
    print(dataset[i].isna().sum())
    dataset[i] = dataset[i].fillna(dataset[i].mean())
    print(dataset[i].isna().sum())
    print("-----------------")
```

```python
dataset[["Abbreviation"]].value_counts()
```

# **Kategorik -> Nümerik dönüşümü**

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
dataset['Abbreviation'] = le.fit_transform(dataset['Abbreviation'])

ohe = preprocessing.OneHotEncoder()
Abbreviation = ohe.fit_transform(dataset.iloc[:,4:5]).toarray()
Abbreviation_df = pd.DataFrame(Abbreviation, columns=list(le.classes_))

dataset = pd.concat([dataset.iloc[:,:4], Abbreviation_df, dataset.iloc[:,5:]], axis=1)
dataset.head().T
```

```python
le = preprocessing.LabelEncoder()
dataset['category'] = le.fit_transform(dataset['category'])

ohe = preprocessing.OneHotEncoder()
category = ohe.fit_transform(dataset.iloc[:,2:3]).toarray()
category_df = pd.DataFrame(category, columns=list(le.classes_))

dataset = pd.concat([dataset.iloc[:,:2], category_df, dataset.iloc[:,3:]], axis=1)
dataset.head(5)
```

# **Makine Öğrenmesi**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
y = dataset[['channel_type']]
x = dataset.drop(columns=['channel_type'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x ,y, test_size=0.25, random_state=0)
```

```python
rfc = RandomForestClassifier(criterion='entropy', max_depth=100)
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)

cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
print(cm)
print(acc)
```
