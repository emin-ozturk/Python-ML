# Kütüphaneler

```python
import seaborn as sns
```

# Veri seti yükleme

```python
tips = sns.load_dataset("tips")
tips
```

# Veri analizi

```python
print(tips.info(), end="\n\n")
print(tips.shape, end="\n\n")
print(tips.isna().sum(), end="\n\n")
print(tips.index, end="\n\n")
```

```python
print(tips["day"].nunique())
print(tips["day"].value_counts())
```

```python
print(tips["time"].nunique())
print(tips["time"].value_counts())
```

# Kırılımlar

```python
tips.groupby("time").agg({"total_bill": ["max", "min", "mean"]})
```

```python
tips.groupby(["sex", "size"]).agg({"tip": ["max", "min", "mean", "count"]})
```

# Sütun birleştirme

```python
agg_df = tips.groupby(["sex", "smoker", "day", "time"]).agg(
        {
            "total_bill": "mean",
            "tip": "mean"
        })
agg_df.head()
```

```python
agg_df.reset_index(inplace = True)
agg_df.head()
```

```python
agg_df.shape
```

```python
agg_df.dropna(axis = 0, inplace = True)
agg_df.shape
```

```python
agg_df
```

```python
agg_df["general_col"] = agg_df[["sex", "smoker", "day", "time"]].agg(lambda x: "_".join(x).upper(), axis = 1)
```

```python
agg_df.head()
```

```python
user = "FEMALE_NO_SAT_DINNER"
agg_df[agg_df["general_col"] == user]
```
