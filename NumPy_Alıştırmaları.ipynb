```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0, 100, size = 20)
y = np.random.randint(0, 100, size = 20)
print(x)
print(y)
```

```python
print(type(x))
print(type(y))
```

```python
plt.figure(figsize = (5, 3))
sns.lineplot(x = x, y = y)
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.scatterplot(x = x, y = y)
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.lineplot(x = x, y = y, color = "green", marker = "*")
plt.show()
```

```python
x1 = np.random.randint(0, 20, 20)
x2 = np.random.randint(0, 20, 20)
y1 = np.random.randint(0, 20, 20)
y2 = np.random.randint(0, 20, 20)

plt.figure(figsize = (5, 3))
sns.lineplot(x = x1, y = y1, color = "blue", label='Grafik 1')
sns.lineplot(x = x2, y = y2, color = "red", label='Grafik 2')

plt.legend()
plt.show()
```

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([6, 7, 8, 9, 10])

a2 = a ** 2
print(f"a2: {a2}")

b2 = b / 2
print(f"b2: {b2}")

c = a2 * b2
print(f"c: {c}")
```
