# Kütüphaneler

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

# Veri yükleme

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
np.random.seed(0)
dataset.sample(5)
```

# Veri silme

```python
clean_dataset = dataset.drop(["PassengerId", "Name", "Ticket"], axis=1)
```

```python
clean_dataset
```

```python
le = preprocessing.LabelEncoder()
clean_dataset["Sex"] = le.fit_transform(clean_dataset["Sex"])
```

# Eksik veri doldurma

```python
clean_dataset.isna().sum()
```

```python
print(clean_dataset.Age.unique())
print(clean_dataset.Cabin.unique())
print(clean_dataset.Embarked.unique())
```

## En çok tekrar eden ile doldurma

```python
print("En çok tekrar eden değerler")
print("Age: ", clean_dataset.Age.value_counts().idxmax())
print("Cabin: ", clean_dataset.Cabin.value_counts().idxmax())
print("Embarked: ", clean_dataset.Embarked.value_counts().idxmax())
```

```python
most_repeat_dataset = clean_dataset.fillna(
    {
        "Age": clean_dataset.Age.value_counts().idxmax(),
        "Cabin": clean_dataset.Cabin.value_counts().idxmax(),
        "Embarked": clean_dataset.Embarked.value_counts().idxmax()
    })
np.random.seed(0)
most_repeat_dataset.sample(5)
```

```python
most_repeat_dataset = pd.get_dummies(most_repeat_dataset, columns=["Cabin", "Embarked"], prefix=["Cabin", "Embarked"])
print(most_repeat_dataset.shape)
```

## Bir önceki veri ile doldurma

```python
bfill_dataset = clean_dataset.fillna(method="bfill")
np.random.seed(0)
bfill_dataset.sample(5)
```

```python
bfill_dataset = pd.get_dummies(bfill_dataset, columns=["Cabin", "Embarked"], prefix=["Cabin", "Embarked"])
print(bfill_dataset.shape)
```

## Eksik verileri silme

```python
drop_dataset = clean_dataset.dropna()
np.random.seed(0)
drop_dataset.sample(5)
```

```python
print(clean_dataset.shape)
print(drop_dataset.shape)
```

```python
drop_dataset = pd.get_dummies(drop_dataset, columns=["Cabin", "Embarked"], prefix=["Cabin", "Embarked"])
print(drop_dataset.shape)
```

## Interpolate

```python
interpolate_dataset = clean_dataset.interpolate(method="index")
# Sayısal olmayan değerler bir sonraki değer ile doldurulur
interpolate_dataset = interpolate_dataset.fillna(method="ffill")

np.random.seed(0)
interpolate_dataset.sample(5)
```

```python
interpolate_dataset = pd.get_dummies(interpolate_dataset, columns=["Cabin", "Embarked"], prefix=["Cabin", "Embarked"])
print(interpolate_dataset.shape)
```

# Makine Öğrenmesi
Farklı yöntemler ile doldurulan verilerin başarı karşılaştırması

## Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
def x_and_y(dataset):
  y = dataset.Survived
  x = dataset.drop(["Survived"], axis=1)
  return x, y
```

```python
datasets = [most_repeat_dataset, bfill_dataset, drop_dataset, interpolate_dataset]
datasets_name = ["most_repeat_dataset", "bfill_dataset", "drop_dataset", "interpolate_dataset"]
for i, dataset in enumerate(datasets):
  x, y = x_and_y(dataset)
  np.random.seed(0)
  X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
  rfc = RandomForestClassifier()
  rfc.fit(X_train, y_train)
  y_pred = rfc.predict(X_test)
  acc = accuracy_score(y_test, y_pred)
  print("{0} : {1}".format(datasets_name[i], acc))
```
