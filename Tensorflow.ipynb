```python
import pandas as pd
import tensorflow as tf
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
y = tf.keras.utils.to_categorical(y)

print("Therefore, our final shape of output feature will be {}".format(y.shape))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.33, random_state=0)
print("Training Input shape\t: {}".format(x_train.shape))
print("Testing Input shape\t: {}".format(x_test.shape))
print("Training Output shape\t: {}".format(y_train.shape))
print("Testing Output shape\t: {}".format(y_test.shape))
```

```python
model = Sequential()
model.add(Dense(258, kernel_initializer = 'uniform', activation='relu', input_dim = 4))
model.add(Dense(128, kernel_initializer = 'uniform', activation='relu'))
model.add(Dense(128, kernel_initializer = 'uniform', activation='relu'))
model.add(Dense(64, kernel_initializer = 'uniform', activation='relu'))
model.add(Dense(3, kernel_initializer = 'uniform', activation='sigmoid'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
model.fit(x_train, y_train, epochs=200, batch_size=8)
y_pred = model.predict(x_test)
```

```python
accuracy = model.evaluate(x_test, y_test)[1]
print("Test doğruluğu:", accuracy)
```
